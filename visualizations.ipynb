# Packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
data = pd.read_csv('dataFullNoFacilities.csv', encoding='utf-8-sig', index_col=0)

apartments = data[data['Product Type'] == 'apartment']
houses = data[data['Product Type'] == 'house']
```

# Visualizations

## Apartments 

### Price Distribution

```python
ax = sns.displot(apartments, x="Price", color='green').set(title='Apartments Price Distribution')
ax.fig.set_figwidth(10)
ax.fig.set_figheight(5.3333)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
```

### Surface Area 

```python
ax = sns.displot(apartments, x="Interior Area", color='blue').set(title='Surface Area Apartments')
ax.fig.set_figwidth(10)
ax.fig.set_figheight(5.3333)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
```

### Rooms Distribution

```python
sns.displot(apartments, x="Room Count", discrete=True, color='red')
```

### Bathrooms Distribution

```python
sns.displot(apartments, x="Bathroom Count", discrete=True, color='Yellow').set(title = 'Apartments Bathroom Distribution')
```

### Material Distribution

```python
monolit_count = apartments['Material_Monolit'].value_counts()[1]
panel_count = apartments['Material_Panel'].value_counts()[1]
stone_count = apartments['Material_Stone'].value_counts()[1]
bricks_count = apartments['Material_bricks'].value_counts()[1]
cassette_count = apartments['Material_cassette'].value_counts()[1]
other_count = apartments['Material_other'].value_counts()[1]
wooden_count = apartments['Material_wooden'].value_counts()[1]
```

```python
labels = 'Monolith', 'Panel', 'Stone', 'Brick', 'Cassette', 'Other', 'Wooden'
sizes = [monolit_count, panel_count, stone_count, bricks_count,
         cassette_count, other_count, wooden_count]
 
explode = (0.2, 0.2, 0.2, 0.2, 2, 2, 2) 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels)
plt.show()
```
